# **Modeling Fuel Efficiency: A Comparative Study of Linear and Ensemble Regression Approaches**

## **Business Problem Statement**

The automotive industry and consumers alike are increasingly focused on fuel efficiency (miles per gallon, MPG) as a key indicator of environmental sustainability and cost-effectiveness. Rising fuel prices, tightening emissions standards, and the shift toward energy-conscious design have made it vital for manufacturers and policymakers to understand which vehicle attributes most influence fuel economy.

However, while automakers collect large volumes of vehicle specification data—engine displacement, horsepower, weight, and model year—the relationships between these factors and actual fuel efficiency remain complex and often nonlinear. Traditional linear methods may oversimplify these dynamics, while modern machine learning approaches promise greater predictive power but less interpretability.

## **Research Objectives**

The primary objective of this research is to explore how different regression modeling techniques—both traditional and ensemble—can be applied to predict automobile fuel efficiency, measured in miles per gallon (MPG).

Through this investigation, the study seeks to:

1. **Quantify** the relationships between key vehicle characteristics (e.g., engine displacement, horsepower, weight, model year) and fuel efficiency.
1. **Compare** the predictive performance of four regression methods:
    * Linear Regression (baseline)
    * Decision Tree Regressor
    * Random Forest Regressor
    * Gradient Boosting Regressor
1. **Evaluate** trade-offs between model interpretability, accuracy, and computational efficiency.
1. **Determine** which modeling approach provides the most balanced performance for predicting MPG in small-to-moderate-sized automotive datasets.
1. **Provide** actionable insights for stakeholders—automotive engineers, data scientists, and consumers—regarding which factors most influence vehicle fuel economy.

## **Research Questions**

This study is guided by the following research questions:

1. **Predictive Accuracy** — Which regression modeling technique provides the highest predictive accuracy for estimating vehicle MPG?
1. **Model Complexity and Overfitting** — How do the models differ in terms of their ability to generalize to unseen data without overfitting?
1. **Feature Influence** — Which vehicle characteristics most strongly impact predicted fuel efficiency across models?
1. **Interpretability vs. Performance** — How does model interpretability change as predictive performance improves from linear to ensemble approaches?
1. **Operational Efficiency** — What are the relative computational costs (fit time and prediction time) of each model, and how might these affect real-world deployment?

## **Methodology and Data Description**

### **Overview of Methodology**

This study employs a **comparative regression modeling approach** to predict vehicle fuel efficiency (measured in miles per gallon, MPG) using the **Auto MPG dataset** from the UCI Machine Learning Repository.
The workflow was designed to balance **statistical rigor, interpretability, and computational fairness** across multiple modeling techniques.

The modeling process followed five key phases:

1. **Data Acquisition and Exploration** – Load and inspect the dataset for completeness, structure, and basic statistical properties.
1. **Data Cleaning and Preprocessing** – Remove irrelevant columns, handle missing or invalid values, and encode categorical variables.
1. **Feature Engineering** – Convert text-based columns to numeric form and create dummy variables for categorical features.
1. **Model Training and Evaluation** – Train four regression models (Linear Regression, Decision Tree, Random Forest, and Gradient Boosting) using identical train/test splits with 5-fold cross-validation for robust model comparison.
1. **Performance Comparison and Interpretation** – Compare models on ( R^2 ), RMSE, MAE, and computation time to evaluate predictive performance, generalization, and efficiency.

### **Dataset Description**

The dataset used in this research, **auto-mpg.csv**, contains specifications of various automobiles produced between **1970 and 1982**.
It consists of **398 observations** and **9 attributes**, one of which (MPG) serves as the target variable.

**Table 1: Auto MPG Data Dictionary**

| **Feature Name** | **Description**                                        | **Data Type** | **Role**              |

### **Data Preprocessing Steps**

To ensure data integrity and model readiness, the following preprocessing steps were executed:

1. **Dropped** the `car name` column due to its non-numeric format and lack of predictive value.
1. **Converted** `horsepower` from string to numeric, coercing invalid entries to `NaN`.
1. **Imputed** missing `horsepower` values using the column mean.
1. **Encoded** the categorical `origin` variable using one-hot encoding (creating `origin_1`, `origin_2`, `origin_3`) and `drop_first=True` parameter to reduce multicollinearity.
1. **Scaled** predictor features for the Linear Regression model using `StandardScaler` to ensure comparability across numeric magnitudes.
1. **Applied** an 80/20 train-test split (`random_state=42`) and 5-fold cross-validation to assess stability and generalization across models.

### **Comparative Analysis**

#### **Linear Regression (Baseline Model)**

Linear Regression, the simplest model, established a baseline for interpretability and computational speed.

* It achieved a **test R² of 0.845**, explaining roughly 85% of the variance in MPG.
* However, it showed the **highest prediction error (RMSE = 2.89)** among all models.
* The **fit time (0.012s)** was negligible, reinforcing its efficiency and suitability for quick, interpretable benchmarks.
  These results suggest that while Linear Regression captures broad linear relationships (e.g., between weight and MPG), it fails to account for nonlinear interactions between variables like horsepower, displacement, and acceleration.

#### **Decision Tree Regressor**

The Decision Tree model improved predictive power over Linear Regression, with a **test R² of 0.876** and lower **test RMSE (2.58)**.

* The model’s **max_depth = 6** and **min_samples_leaf** tuning helped prevent severe overfitting.
* Its interpretability makes it useful for identifying feature thresholds (e.g., “weight below 2,000 lbs → higher MPG”).
  However, despite this improvement, the Decision Tree’s single-structure nature limited its generalization compared to ensemble approaches.

#### **Random Forest Regressor**

Random Forest achieved the **best overall performance**, with the **highest test R² (0.915)** and **lowest test RMSE (2.14)**.

* It showed strong agreement between training and test performance, indicating effective variance reduction.
* The selected hyperparameters (**max_depth=10**, **max_features='sqrt'**, and **min_samples_leaf≈2**) allowed it to generalize well while maintaining interpretability through aggregated feature importance scores.
* While its **fit time (9.12s)** was higher than other models, the improvement in accuracy and generalization justified the computational cost.
  This suggests Random Forest offers the best balance between **bias, variance, and predictive reliability** for this dataset.

#### **Gradient Boosting Regressor**

Gradient Boosting produced the **highest cross-validation R² (0.873)** but slightly **lower test R² (0.895)** compared to Random Forest.

* The best configuration (**learning_rate=0.05**, **max_depth=3**, **n_estimators=200–300**) emphasizes gradual error correction over aggressive fitting.
* It achieved competitive performance with **RMSE (2.37)** and **MAE (1.76)** with shorter training time (2.65s).
  These results show that Gradient Boosting generalizes consistently across folds, although it can underperform slightly on individual splits due to its sensitivity to learning rate and estimator count.
  In practice, Gradient Boosting is a **strong, computationally efficient alternative** to Random Forest when fine-tuned properly.

### **Discussion and Interpretation**

The results demonstrate a clear progression in model capability as complexity increases from **Linear Regression → Decision Tree → Ensemble Methods**.

* **Linear Regression** captures broad linear patterns but overlooks nonlinearities.
* **Decision Tree** models introduce flexibility but are prone to instability when used alone.
* **Random Forest** mitigates these weaknesses by averaging across trees, reducing variance without losing interpretability.
* **Gradient Boosting**, while highly accurate and efficient, requires more careful tuning and may underperform slightly on unseen splits if hyperparameters are not fully optimized.

From a practical standpoint:

* **Manufacturers** could use models like Random Forest to estimate expected fuel efficiency based on design parameters early in product development.
* **Consumers** could benefit from simplified models (like Decision Trees) for transparent, interpretable predictions when comparing vehicles.
* **Data scientists** can view this comparison as a baseline benchmark illustrating how ensemble methods enhance predictive performance without drastically increasing computational cost.

### **Summary of Findings**

## **Conclusions and Recommendations**

### **Conclusions**

This study, *“Modeling Fuel Efficiency: A Comparative Study of Linear and Ensemble Regression Approaches,”* set out to identify which vehicle characteristics most influence fuel economy and to compare the performance of several regression models in predicting miles per gallon (MPG).

Using the **Auto MPG dataset**, four models—**Linear Regression**, **Decision Tree**, **Random Forest**, and **Gradient Boosting**—were developed, validated, and compared using consistent cross-validation and test procedures.

The key conclusions are as follows:

1. **Ensemble methods outperform traditional regression models** in predicting MPG.

   * Random Forest achieved the highest overall accuracy with a test (R^2 = 0.915) and lowest RMSE (2.14).
   * Gradient Boosting provided nearly comparable performance with the fastest ensemble training time and strongest average cross-validation stability.

1. **Vehicle weight, displacement, and horsepower are the most influential predictors** of fuel efficiency.
   Heavier and more powerful vehicles consistently exhibited lower MPG values across all models, confirming well-established engineering and economic trends.

1. **Model complexity correlates with accuracy but reduces interpretability.**
   While Random Forest and Gradient Boosting offered substantial accuracy gains, they require more computation and lack the transparency of simpler models like Linear Regression or a tuned Decision Tree.

1. **Cross-validation proved essential for robust evaluation.**
   Ensemble models exhibited consistent (R^2) scores across folds, confirming strong generalization, while Decision Tree results varied more due to sensitivity to small changes in training data.

Overall, the research demonstrates that **ensemble learning methods—particularly Random Forest—provide the best balance between predictive performance and model robustness** for estimating vehicle fuel efficiency in small-to-medium-sized datasets.

A residual diagnostics study was conducted for each model (see Appendix A). Residual plots confirmed homoscedasticity and lack of bias, supporting the validity of model assumptions and generalization integrity.

### **Recommendations**

````python
pipelines = {
    # Best suited for simple regression tasks where data distribution is expected to follow a linear trend.
    # This model provides speed and interpretability ideal for understanding the impact of each feature.
    # Linear Regression models may underperform on non-linear relationships unless polynomial or 
    # interaction terms are included.
    "Linear Regression": Pipeline([
        ("scaler", StandardScaler()),
        ("model", LinearRegression())
    ]),
    # Decision Tree Regressor splits data into branches based on decision nodes which are highly 
    # interpretable and easy to visualize.  These models are well-suited for numerical and categorical data
    # without the need for scaling.  This model is prone to overfitting unless constraints such as tree
    # depth and minimum sample splits are applied.  Use when decisions need to be modeled explicitly and 
    # relationships between features need to be clearly understood.
    "Decision Tree": Pipeline([
        ("model", DecisionTreeRegressor(random_state=RANDOM_STATE))
    ]),
    # This model reduces the risk of overfitting common to individual decision trees by averaging multiple 
    # trees' results.  Random Forest can handle a large number of features and can be used for feature 
    # importance evaluation.  The model is effective for regression tasks, where the goal is to predict a
    # continuous variable.
    "Random Forest": Pipeline([
        ("model", RandomForestRegressor(random_state=RANDOM_STATE, n_jobs=-1))
    ]),
    # Gradient Boosting Regressor builds sequential models sequentially, each new model corrects errors 
    # made by the previous ones.  This model is effective for complex regression tasks involving 
    # non-linear relationships.  The model can be fine-tuned to avoid overfitting through parameters such
    # as learning rate and number of trees.
    "Gradient Boosting": Pipeline([
        ("model", GradientBoostingRegressor(random_state=RANDOM_STATE))
    ]),
}
````

#### **Define Modeling Functions**

##### **Function to Calculate Cross-Validation Scores**

````python
def _cv_summary(estimator, X, y, cv):
    """Return mean/stdev for r2, rmse, mae across CV folds (rmse/mae flipped to positive)."""
    out = {}
    for k, s in scorers.items():
        scores = cross_val_score(estimator, X, y, scoring=s, cv=cv, n_jobs=-1)
        if k in ("rmse", "mae"):
            scores = -scores  # invert sign to get positive errors
        out[f"cv_{k}_mean"] = float(np.mean(scores))
        out[f"cv_{k}_std"] = float(np.std(scores))
    return out
````

##### **Function to calculate Root Mean Squared Error (RMSE) of Models**

````python
def _rmse(a, b): return float(np.sqrt(mean_squared_error(a, b)))
````

##### **Function to Evaluate Model and Return Results**

````python
def evaluate_model(name, pipe, grid, X_train, y_train, X_test, y_test, return_artifacts=False):
    """Fit (with optional GridSearch), gather CV + holdout metrics, and timing."""
    t0 = time.time()
    if grid:
        # multi-metric CV; refit by best r2
        gs = GridSearchCV(
            pipe, grid, cv=cv, scoring=scorers, refit="r2", n_jobs=-1, verbose=0
        )
        gs.fit(X_train, y_train)
        best = gs.best_estimator_
        search_time = time.time() - t0
        cv_stats = _cv_summary(gs.best_estimator_, X_train, y_train, cv)
        chosen_params = gs.best_params_
    else:
        pipe.fit(X_train, y_train)
        best = pipe
        search_time = time.time() - t0
        cv_stats = _cv_summary(best, X_train, y_train, cv)
        chosen_params = {}

    # Predict + test metrics
    t1 = time.time()
    ytr_pred = best.predict(X_train)
    yte_pred = best.predict(X_test)
    pred_time = time.time() - t1

    row = {
        "model": name,
        # CV summary
        **cv_stats,
        # Train metrics
        "train_r2": float(r2_score(y_train, ytr_pred)),
        "train_rmse": _rmse(y_train, ytr_pred),
        "train_mae": float(mean_absolute_error(y_train, ytr_pred)),
        # Test metrics
        "test_r2": float(r2_score(y_test, yte_pred)),
        "test_rmse": _rmse(y_test, yte_pred),
        "test_mae": float(mean_absolute_error(y_test, yte_pred)),
        # Timing
        "fit_time_s": float(search_time),
        "predict_time_s": float(pred_time),
        # Params
        "best_params": chosen_params,
    }
    
    if return_artifacts:
        artifacts = {
            "name": name,
            "estimator": best,
            "y_train_pred": ytr_pred,
            "y_test_pred": yte_pred,
        }
        return row, artifacts
    else:
        return row
````

````python
rows = []
artifacts = []

for name, pipe in pipelines.items():
    r, a = evaluate_model(name, pipe, param_grids[name], X_train, y_train, X_test, y_test, return_artifacts=True)
    rows.append(r)
    artifacts.append(a)
````

#### **Residual Analysis**

Residual plots display the difference between actual and predicted MPG values for each model on both training and test sets. The x-axis shows predicted MPG values, while the y-axis shows residuals (actual minus predicted). Ideally, residuals should be randomly scattered around zero with no discernible patterns, indicating that the model has captured all systematic relationships in the data. Patterns such as curves, funneling (changing variance), or clustering suggest the model is missing important relationships or making systematic errors. Comparing training and test residual plots helps identify overfitting—if training residuals are much smaller and more randomly distributed than test residuals, the model may have memorized training data rather than learning generalizable patterns.

##### **Reusable Function to Plot Residuals**

````python
def plot_residuals(y_true, y_pred, title):
    residuals = y_true - y_pred
    fig, axes = plt.subplots(1, 2, figsize=(12,5))
    sns.scatterplot(x=y_pred, y=residuals, ax=axes[0])
    axes[0].axhline(0, color='red', linestyle='--')
    axes[0].set_title(f'Residuals vs Predicted ({title})')
    sns.histplot(residuals, kde=True, ax=axes[1])
    axes[1].set_title(f'Residual Distribution ({title})')
    plt.show()
````

##### **Generate Residual Plots for All Models**

**Linear Regression** exhibits the most problematic residual patterns. The training plot shows a clear funnel shape with increasing variance at higher predicted MPG values, and a notable cluster of large positive residuals (10-15 MPG errors) for low-efficiency vehicles. The test set confirms this heteroscedasticity, indicating the linear model fails to capture non-linear relationships in the data. The asymmetric residual distribution (skewed positive) suggests systematic underprediction for certain vehicle types.

**Decision Tree** shows improvement over linear regression with more randomly scattered residuals and reduced variance. However, the training residuals are noticeably tighter than test residuals, and both distributions show slight positive skew. The presence of a few extreme outliers (>10 MPG error) in the test set indicates sensitivity to individual split decisions, consistent with decision tree instability.

**Random Forest** demonstrates the best residual characteristics among all models. Training residuals are well-centered around zero with minimal spread, and test residuals maintain consistent variance across all predicted values (homoscedasticity). The residual distributions are approximately normal with no systematic patterns, indicating the model has successfully captured the underlying relationships. The similar residual behavior between training and test sets confirms good generalization without overfitting.

**Gradient Boosting** produces excellent results comparable to Random Forest. Both training and test residuals are tightly clustered around zero with symmetric, near-normal distributions. The residual variance is remarkably consistent across the prediction range, and the lack of patterns suggests the model has extracted all systematic information from the features. The slightly narrower residual distribution compared to Random Forest reflects its sequential error-correction approach.
Overall, the residual analysis confirms that ensemble methods (Random Forest and Gradient Boosting) successfully address the limitations of simpler approaches, with residual patterns that satisfy the key assumptions for reliable regression modeling.

### **SHAP (SHapley Additive exPlanations) Analysis**

SHAP analysis provides model-agnostic interpretability by quantifying each feature's contribution to individual predictions using principles from cooperative game theory. Three complementary visualization types are presented for each model to examine feature importance from different analytical perspectives.

#### **SHAP Visualization Types**

**Global Feature Importance (Bar Plot)** — Identifies which features matter most across all predictions. Bar length represents the average absolute impact each feature has on MPG predictions. Longer bars indicate more influential features. For example, if weight has the longest bar, it contributes most to explaining MPG variation across all vehicles, regardless of direction of effect.

**Feature Effects and Direction (Beeswarm Plot)** — Shows both which features matter and *how* they affect predictions. Each dot represents one vehicle. Horizontal position indicates impact magnitude (positive = increases MPG, negative = decreases MPG). Color indicates feature value (red = high, blue = low). For weight, red dots (heavy vehicles) cluster left (negative impact) while blue dots (light vehicles) cluster right (positive impact), confirming the negative weight-MPG relationship.

**Dependence Plot for Top Feature (Scatter Plot)** — Examines the detailed relationship between one feature and predictions. The x-axis shows actual feature values, y-axis shows SHAP impact, and color indicates a potentially interacting feature. This reveals whether relationships are linear or non-linear and identifies interaction effects. For example, a weight dependence plot might show that MPG decreases non-linearly with weight, with stronger effects in high-horsepower vehicles.

#### **SHAP Reusable Functions**

The following functions are used to generate SHAP plots for the models.

````python
def get_final_estimator(artifact):
    # your pipelines use .named_steps['model']
    return artifact["estimator"].named_steps.get("model", artifact["estimator"])

def shap_explainer_for(artifact, X_train, model_name):
    est = get_final_estimator(artifact)
    if model_name in ["Decision Tree", "Random Forest", "Gradient Boosting"]:
        return shap.TreeExplainer(est), X_train  # use raw X_train used by the pipeline
    elif model_name == "Linear Regression":
        # Use the pipeline to transform X for a correct background
        X_train_scaled = artifact["estimator"].named_steps["scaler"].transform(X_train)
        return shap.LinearExplainer(est, X_train_scaled), X_train_scaled
    else:
        # Fallback: KernelExplainer (slower); try to avoid
        X_bkg = shap.sample(X_train, 200, random_state=42)
        return shap.KernelExplainer(artifact["estimator"].predict, X_bkg), X_train
````

#### **Generate SHAP Plots for All Models**

##### **Comparative Interpretation**

SHAP plots are presented for all four models to validate consistency of findings across modeling approaches. Agreement on top features (weight, horsepower, displacement) confirms these represent genuine physical relationships rather than model artifacts. Differences in dependence plot shapes demonstrate why ensemble methods outperform linear models—they capture non-linear relationships and feature interactions that linear assumptions cannot represent. This analysis confirms that weight reduction and engine efficiency improvements are the most effective levers for enhancing fuel economy.

````python
for i, row in enumerate(rows):
    name = row["model"]
    print(f"\n=== SHAP for {name} ===")

    explainer, X_bg = shap_explainer_for(artifacts[i], X_train, name)

    if isinstance(explainer, shap.explainers.Linear):
        X_bg = pd.DataFrame(X_bg, columns=X_train.columns, index=X_train.index)

    shap_values = explainer(X_bg)

    # --- Figure D#a: Bar Plot ---
    print(f"\n\033[1mFigure {figure_index} — {name}: Global Feature Importance\033[0m\n")
    shap.plots.bar(shap_values, max_display=10, show=False)
    plt.gcf().axes[0].set_title(f"Figure {figure_index} — {name}: Global Feature Importance", fontsize=12, pad=15)
    figure_index += 1
    plt.tight_layout()
    plt.show()

    # --- Figure D#b: Beeswarm ---
    print(f"\n\033[1mFigure {figure_index} — {name}: Feature Effects and Direction\033[0m\n")
    shap.plots.beeswarm(shap_values, max_display=15, show=False)
    plt.gcf().axes[0].set_title(f"Figure {figure_index} — {name}: Feature Effects and Direction", fontsize=12, pad=15)
    figure_index += 1
    plt.tight_layout()
    plt.show()

    # --- Figure D#c: Dependence Plot (Top Feature) ---
    # Automatically pick the top feature by mean |SHAP|
    print(f"\n\033[1mFigure {figure_index} — {name}: Dependence Plot for Top Feature\033[0m\n")
    mean_abs_shap = np.abs(shap_values.values).mean(axis=0)
    top_idx = mean_abs_shap.argsort()[-1]
    top_feat = shap_values.feature_names[top_idx]
    shap.plots.scatter(shap_values[:, top_idx], color=shap_values, show=False)
    plt.gcf().axes[0].set_title(f"Figure {figure_index} — {name}: Dependence Plot for {top_feat}", fontsize=12, pad=15)
    figure_index += 1
    plt.tight_layout()
    plt.show()
````
